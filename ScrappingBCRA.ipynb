# Scrapping **BCRA**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
url = 'http://www.bcra.gov.ar/SistemasFinancierosYdePagos/Emisoras_tarjetas_credito_compra.asp'
pagina = requests.get(url)
```

```python
# parser-lxml = Change html to Python friendly format
# Obtain page's information
soup = BeautifulSoup(pagina.text, 'lxml')
soup
```

```python
# Obtain information from tag <table>
table1 = soup.find("table", {"class": "table table-BCRA table-bordered table-hover table-responsive post-pagina-interior-txtchico"})
table1
```

```python
# Obtain every title of columns with tag <th>
headers = []

for i in table1.find_all('tr'):
  for j in table1.find_all('td'):
    title = j.text
    headers.append(title)
print(headers)
# table1
```

```python
# Create a for loop to fill mydata
# for j in table1.find_all('tr')[2:]:
for j in table1.find_all('tr'):  
 row_data = j.find_all('td')
 row = [i.text for i in row_data]
 print(row)
```

```python
cabe = ['Código emisora', 'Entidad emisora', 'CUIT', 'Nombre de la tarjeta', 'Calle y Nº', 'Localidad', 'Provincia', 'Código postal', 'Teléfono']
tarjetas = pd.DataFrame(columns = cabe)
tarjetas
```

```python
for j in table1.find_all('tr')[2:]:
 row_data = j.find_all('td')
 row = [i.text for i in row_data]
 length = len(tarjetas)
 tarjetas.loc[length] = row
 
```
